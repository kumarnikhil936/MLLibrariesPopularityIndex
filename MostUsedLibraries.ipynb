```
import urllib.request
from time import sleep
import pandas as pd
import bar_chart_race as bcr
from tqdm import tqdm
import re
import cProfile
```

```
libraries = ['tensorflow', 'torch', 'theano', 'pyspark', 'cv2', 'keras', 'sklearn', 'pandas', 'numpy', 'scipy', 'matplotlib', 'seaborn', 'nltk', 'cython']

expr = re.compile(r'([0-9,]+) (available|commit)')

start_year = 2000
end_year = 2020

language="Python"
```

```
df = []

for year in range(start_year, end_year + 1):
    year_overview = {'date': f'{str(year)}-01-01',}

    for search_term in libraries:
        year_overview[search_term] = 0

    for search_term in libraries:
        while True:
            try:
                url_base = f"https://github.com/search?q={search_term}%28+language%3A{language}+type%3Acommits+committer-date%3A<{str(year)}-01-01&type=commits"
                fp = urllib.request.urlopen(url_base)
                byte_html = fp.read()
                html_string = byte_html.decode("utf8")
                fp.close()

                if "You have triggered an abuse detection mechanism." in html_string:
                    print("Abuse detection mechanism detected. Sleep for 10 minutes!")
                    sleep(60 * 10)
                    continue

                substring = html_string[55000:56000].replace('\n', '')

                try: 
                    val = next(re.finditer(expr, substring)).groups()[0]
                    val = int(val.replace(',', ''))
                except StopIteration:
                    val = 0

                year_overview[search_term] = val
                break

            except Exception as e:
                print("API throttles more than 30 times a minute **", e, "**, when querying for the term: ", search_term, " ** Sleeping for 60 seconds!")
                sleep(60)
                continue

    df.append(year_overview)
    print("\nCollection done for year:", year)
```

```
df = pd.DataFrame(df)
df = df.set_index('date')
df.to_csv(f"libraries_{start_year}_{end_year}.csv", index=True)
```
